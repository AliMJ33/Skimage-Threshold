Thresholding

```python
import matplotlib.pyplot as plt
%matplotlib inline
import skimage.data as data

img = data.camera()

thresh = 127

output1 = img > thresh
output2 = img <= thresh

output = [img, output1, output2]

for i in range(3):
    plt.imshow(output[i], cmap='gray')
    plt.axis('off')
    plt.show()
```

```python
from skimage.filters import threshold_otsu

thresh = threshold_otsu(img)
binary = img > thresh

plt.subplots_adjust(wspace=0.75, hspace=0.75)

plt.subplot(1,3,1)
plt.imshow(img, cmap='gray')
plt.axis('off')

plt.subplot(1,3,2)
plt.hist(img.ravel(), bins=256, range=[0, 255])
plt.xlim([0, 255])

plt.subplot(1,3,3)
plt.imshow(binary, cmap='gray')
plt.axis('off')

plt.show()
```

```python
from skimage.filters import threshold_local

img = data.page()

thresh = threshold_otsu(img)
otsu = img > thresh

adaptive = threshold_local(img, block_size=5, offset=40)

output = [img, otsu, adaptive]

for i in range(3):
    plt.imshow(output[i], cmap='gray')
    plt.axis('off')
    plt.show()
```
